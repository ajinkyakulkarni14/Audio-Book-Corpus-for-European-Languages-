```python
class myWidget(object):
    def __init__(self,sound_files_annotate,sample_rate):
        self.sound_files_annotate=sound_files_annotate
        self.cnt=1
        self.sample_rate=sample_rate
    
    def show(self):
        child_style = {
            'margin': '3px',
            'font-size': 'large',
            'color': 'black',
            'width': '900px',
            'height': '25px'
        }
        i=self.cnt
        display(self.sound_files_annotate[i]["sample_no"])
        #print(self.cnt)
        self.string1 = widgets.TextWidget(description="Prev:",value=sound_files_annotate[i-1]["target"])
        self.string2 = widgets.TextWidget(description="Current:",value=sound_files_annotate[i]["target"])
        self.string3 = widgets.TextWidget(description="Next:",value=sound_files_annotate[i+1]["target"])
        self.string1.set_css(child_style)
        self.string2.set_css(child_style)
        self.string3.set_css(child_style)
        display(self.string1)
        display(self.string2)
        display(self.string3)
        self.button_prev = widgets.ButtonWidget(description = "Prev")
        self.button_next= widgets.ButtonWidget(description = "Next")
        self.button_close= widgets.ButtonWidget(description = "Close")
        self.button_save= widgets.ButtonWidget(description = "Save")
        display(self.button_prev, self.button_next,self.button_close,self.button_save)
        self.button_prev.on_click(self.on_prev_clicked)
        self.button_next.on_click(self.on_next_clicked)
        self.button_close.on_click(self.on_close_clicked)
        self.button_save.on_click(self.on_save_clicked)
        IPython.display.clear_output()
        self.audio=Audio(self.sound_files_annotate[i]["data"],rate=self.sample_rate)
        self.audio.reload()
        display(self.audio)
        
    
    def on_save_clicked(self,b):
        self.update_text()
        fp=codecs.open('book_annotation.txt','w','utf-8')
        for i in self.sound_files_annotate:
            fp.write(str(self.sound_files_annotate[i]['sample_no'])+":"+str(self.sound_files_annotate[i]['min'])+"-")
            fp.write(str(self.sound_files_annotate[i]['max'])+":("+sound_files_annotate[i]["target"]+")\n")
        fp.close()
        #self.on_close_clicked()
        
    def on_close_clicked(self,b):
        IPython.display.clear_output()
        self.string1.close()
        self.string2.close()
        self.string3.close()
        self.button_prev.close()
        self.button_next.close()
        self.button_close.close()
        self.button_save.close()
        
    def update(self):
        IPython.display.clear_output()
        self.string1.close()
        self.string2.close()
        self.string3.close()
        self.button_prev.close()
        self.button_next.close()
        self.button_close.close()
        self.button_save.close()
        self.show()
        #display(self.cnt)
        
    def on_prev_clicked(self, b):
        self.update_text()
        self.cnt -= 1
        if self.cnt-1 < 0:
            self.cnt=1
        self.update()

    def on_next_clicked(self, b):
        self.update_text()
        self.cnt += 1
        self.update()
    
    def update_text(self):
        self.sound_files_annotate[self.cnt-1]['target']=self.string1.value
        self.sound_files_annotate[self.cnt]['target']=self.string2.value
        self.sound_files_annotate[self.cnt+1]['target']=self.string3.value
```

```python
dir_path="/home/ajinkyak/Workspace/Speech/Speech_Anotation/New_project/DIE VERWANDLUNG/die_verwandlung/1/"
sound_files_annotate=collections.defaultdict(dict)
os.chdir(dir_path)
for file in glob.glob("*.txt"):
    annot_file=dir_path+file
fp=codecs.open(annot_file,"r","utf-8")
for f in fp:
    sound_files_annotate[int(f.split(":")[0].strip())]["target"]= f.split(":")[1].strip()
    sound_files_annotate[int(f.split(":")[0].strip())]["sample_no"]= f.split(":")[1].strip()
fp.close()
```
